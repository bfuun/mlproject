```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
```

```python
svrmodel = SVR()
```

```python
param_gridsvr = {
    'C' : [0.1, 1, 10],
    'epsilon' : [0.1, 0.2, 0.5],
    'kernel' : ['linear', 'poly', 'rbf'], 
    'degree' : [2, 3, 4]
}
```

```python
gridsvr = GridSearchCV(svrmodel, param_gridsvr)
```
